```python
import yfinance as yf
import pandas as pd
import numpy as np
import pandas_ta as ta
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_curve
from matplotlib.lines import Line2D
```

```python
df = yf.download("AAPL", start="2020-01-01", end="2024-01-01", interval='1d')
```

```python
if isinstance(df.columns, pd.MultiIndex):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
```

```python
df = df.rename(columns={
    'Open_AAPL': 'Open',
    'High_AAPL': 'High',
    'Low_AAPL': 'Low',
    'Close_AAPL': 'Close',
    'Volume_AAPL': 'Volume'
})

# 3. Reset index to move 'Date' from index to a column (optional but helpful)
df = df.reset_index()

# 4. Now add technical indicators—all at once, on a clean DataFrame!
df.ta.sma(length=20, append=True)
df.ta.ema(length=20, append=True)
df.ta.rsi(length=14, append=True)
df.ta.macd(append=True)
# Add more indicators here as needed (atr, bbands, obv, etc.)
```

```python
# If not already reset after loading/renaming, do it here
df = df.reset_index()

# Drop all rows with any NA values (after indicators are added)
df = df.dropna()

# Remove duplicate dates if any
df = df.drop_duplicates(subset='Date')

# Remove any rows where price columns are non-positive
df = df[(df[['Close', 'Open', 'High', 'Low']] > 0).all(axis=1)]

# Plot close price
plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['Close'], label='Close Price')
plt.title('Close Price Over Time')
plt.legend()
plt.show()

# Plot volume
plt.figure(figsize=(12,3))
plt.plot(df['Date'], df['Volume'], label='Volume', color='orange')
plt.title('Volume Over Time')
plt.legend()
plt.show()

# Outlier detection for Close price
close_mean = df['Close'].mean()
close_std = df['Close'].std()
outliers = df[(df['Close'] > close_mean + 3*close_std) | (df['Close'] < close_mean - 3*close_std)]
print("Extreme close price outliers:\n", outliers)

# Check for missing trading dates
date_range = pd.date_range(df['Date'].min(), df['Date'].max(), freq='B')  # 'B' = business day
missing_dates = set(date_range) - set(df['Date'])
print(f"Missing trading dates: {sorted(missing_dates)}")
```

```python
# Remove zero-volume days if any
if (df['Volume'] == 0).any():
    print(df[df['Volume'] == 0])
    df = df[df['Volume'] > 0]

# Extreme high volume days
vol_mean = df['Volume'].mean()
vol_std = df['Volume'].std()
extreme_volume = df[df['Volume'] > vol_mean + 3 * vol_std]
print("Extreme high volume days:\n", extreme_volume)

# Quick data overview
print(df.info())
print(df.describe())
print(df.head())

# Price & Moving Averages
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['SMA_20'], label='SMA 20')
plt.plot(df['Date'], df['EMA_20'], label='EMA 20')
plt.legend()
plt.title('Price and Moving Averages')
plt.tight_layout()
plt.show()

# Trading Volume
plt.figure(figsize=(14, 4))
plt.bar(df['Date'], df['Volume'], label='Volume')
plt.title('Trading Volume')
plt.tight_layout()
plt.show()

# RSI
plt.figure(figsize=(14, 4))
plt.plot(df['Date'], df['RSI_14'], label='RSI 14')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('RSI Over Time')
plt.legend()
plt.tight_layout()
plt.show()

# Daily Return Distribution
df['Return_1D'] = df['Close'].pct_change()
plt.figure(figsize=(8, 4))
sns.histplot(df['Return_1D'].dropna(), bins=50)
plt.title('Daily Return Distribution')
plt.tight_layout()
plt.show()
```

```python
# Check data types (for EDA/debugging)
print(df.dtypes)

# Correlation heatmap for all numeric features
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f")
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# Set threshold for movement labeling
threshold = 0.001

# Calculate next-day return as target
df['Target'] = (df['Close'].shift(-1) - df['Close']) / df['Close']

# Assign movement labels
df['Behavior'] = np.select(
    [df['Target'] > threshold, df['Target'] < -threshold],
    ['UP', 'DOWN'],
    default='SIDEWAYS'
)

# Drop last row (no next-day close for target)
df = df.dropna(subset=['Target'])

# Show label distribution
print(df['Behavior'].value_counts())
print(df[['Close', 'Target', 'Behavior']].tail())
```

```python
# Safety: Check not empty
assert df.shape[0] > 0, "DataFrame is empty after cleaning!"

# Remove any rows with missing plotting data
df_plot = df.dropna(subset=['Date', 'Close', 'Behavior'])

# Sort by date for robust plotting (usually already sorted)
df_plot = df_plot.sort_values('Date')

# Color mapping for labels
color_map = {'UP': 'green', 'DOWN': 'red', 'SIDEWAYS': 'blue'}
colors = df_plot['Behavior'].map(color_map)

# Plot close price and color-coded behaviors
plt.figure(figsize=(16, 7))
plt.plot(df_plot['Date'], df_plot['Close'], color='black', label='Close Price', alpha=0.5)
plt.scatter(df_plot['Date'], df_plot['Close'], c=colors, label='Behavior', s=30)

# Custom legend (imported at top)
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='UP', markerfacecolor='green', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='DOWN', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='SIDEWAYS', markerfacecolor='blue', markersize=10),
]
plt.legend(handles=legend_elements, title='Behavior')

plt.title('Stock Close Price with Labeled Behaviors')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()
```

```python
features = ['SMA_20', 'EMA_20', 'RSI_14', 'MACD_12_26_9', 'MACDh_12_26_9', 'MACDs_12_26_9', 'Return_1D']
X = df[features]
y = df['Behavior']

# 2. Encode target labels as Pandas Series with matching index
le = LabelEncoder()
y_encoded = pd.Series(le.fit_transform(y), index=y.index)

# 3. Drop NA from features and align target (label-based)
X = X.dropna()
y_encoded = y_encoded.loc[X.index]

# 4. Train-test split (no shuffle for time series)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, shuffle=False
)

# 5. Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6. Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# 7. Feature importances
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10,5))
plt.title("Feature Importance (Random Forest)")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), [X_train.columns[i] for i in indices], rotation=45)
plt.tight_layout()
plt.show()

# 8. Print importances
for i in indices:
    print(f"{X_train.columns[i]}: {importances[i]:.4f}")
```

```python
# Remove 'SIDEWAYS', encode labels, and prepare features
df_bin = df[df['Behavior'].isin(['UP', 'DOWN'])].copy()
le_bin = LabelEncoder()
df_bin['Behavior_num'] = le_bin.fit_transform(df_bin['Behavior'])

# Drop NA
df_bin_clean = df_bin.dropna(subset=features + ['Behavior_num'])

# Features/labels
X_bin = df_bin_clean[features]
y_bin = df_bin_clean['Behavior_num']

# Split (no shuffle for time series)
Xb_train, Xb_test, yb_train, yb_test = train_test_split(X_bin, y_bin, test_size=0.2, shuffle=False)

# Train Random Forest
rf_bin = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_bin.fit(Xb_train, yb_train)
y_bin_pred = rf_bin.predict(Xb_test)

# Evaluation
print("Binary Accuracy:", accuracy_score(yb_test, y_bin_pred))
print(classification_report(yb_test, y_bin_pred, target_names=le_bin.classes_))
print(confusion_matrix(yb_test, y_bin_pred))

print("Training set class distribution:")
print(np.bincount(yb_train))
print("Test set class distribution:")
print(np.bincount(yb_test))
print("Label mapping (index: label):")
print({i: le_bin.classes_[i] for i in range(len(le_bin.classes_))})

# Per-feature boxplots
for feat in features:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df_bin_clean['Behavior_num'], y=df_bin_clean[feat])
    plt.title(f'{feat} distribution by label (0={le_bin.classes_[0]}, 1={le_bin.classes_[1]})')
    plt.show()
```

```python
# Add lag features
df['Return_1D_lag1'] = df['Return_1D'].shift(1)
df['Return_1D_lag2'] = df['Return_1D'].shift(2)

# Technical indicators (add to df, not df_bin_clean)
df.ta.atr(length=14, append=True)
df.ta.bbands(length=20, std=2, append=True)
df.ta.obv(append=True)
df['Volume_Change'] = df['Volume'].pct_change()

# Optional: Manual BBands/OBV if you specifically want those
# df['BB_upper'] = ...
# df['OBV_manual'] = ...

# Now create the binary DataFrame and drop NA
df_bin = df[df['Behavior'].isin(['UP', 'DOWN'])].copy()
le_bin = LabelEncoder()
df_bin['Behavior_num'] = le_bin.fit_transform(df_bin['Behavior'])

# Update your features list!
features = [
    'SMA_20', 'EMA_20', 'RSI_14', 'MACD_12_26_9', 'MACDh_12_26_9',
    'MACDs_12_26_9', 'Return_1D', 'Return_1D_lag1', 'Return_1D_lag2',
    'ATRr_14', 'BBL_20_2.0', 'BBM_20_2.0', 'BBU_20_2.0', 'BBB_20_2.0',
    'OBV', 'Volume_Change'
]
df_bin_clean = df_bin.dropna(subset=features + ['Behavior_num'])

# Build model inputs
X_bin = df_bin_clean[features]
y_bin = df_bin_clean['Behavior_num']

# Split (no shuffle for time series)
Xb_train, Xb_test, yb_train, yb_test = train_test_split(X_bin, y_bin, test_size=0.2, shuffle=False)

# Train classifier
rf_bin = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_bin.fit(Xb_train, yb_train)
y_bin_pred = rf_bin.predict(Xb_test)

# Evaluate
print("Binary Accuracy:", accuracy_score(yb_test, y_bin_pred))
print(classification_report(yb_test, y_bin_pred, target_names=le_bin.classes_))
print(confusion_matrix(yb_test, y_bin_pred))
```
